```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data
df = pd.read_excel("C:\\Users\\HUAWEI\\Downloads\\Online Retail.xlsx")
df.info()
```

```python
#Looking at the first five rows
df.head()
```

```python
#Dropping rows with no customer ID
df = df.dropna(subset=['CustomerID'])
df.info()
```

```python
#Creating a column for invoice month. This will be used later to create a cohort month
import datetime as dt

#Creating a function for month
def get_month(x):
    return dt.datetime(x.year, x.month, 1)


#Applying the function below and looking at the last 5 rows
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)
df.tail()
```

```python
#Creating a column index with the minimum invoice date which would represent the first time a customer was acquired

df['Cohort Month'] = df.groupby('CustomerID')['InvoiceMonth'].transform('min')
df.tail(20)
```

```python
#Creating a date element function to get a series for subtraction

def get_date_elements(data, column):
    day = data[column].dt.day
    month = data[column].dt.month
    year = data[column].dt.year
    return day, month, year 
```

```python
#Getting date element for the cohort and invoice columns

_,Invoice_month,Invoice_year = get_date_elements(df, 'InvoiceMonth')
_,Cohort_month,Cohort_year = get_date_elements(df, 'Cohort Month')
```

```python
#Checking the series
Invoice_month[:10]
```

```python
Cohort_year[:10]
```

```python
#Creating a cohort index that will show when a customer was acquired

year_diff = Invoice_year - Cohort_year
month_diff = Invoice_month - Cohort_month
df['CohortIndex'] = year_diff*12+month_diff+1

df.head()
```

```python
df.tail()
```

```python
#Counting the Customer ID by grouping by cohort month and cohort index

cohort_data = df.groupby(['Cohort Month','CohortIndex'])['CustomerID'].apply(pd.Series.nunique)
cohort_data
```

```python
#Using the reset index function to put the output above in a table

cohort_data = df.groupby(['Cohort Month','CohortIndex'])['CustomerID'].apply(pd.Series.nunique).reset_index()
cohort_data
```

```python
#Creating a pivot table

cohort_table = cohort_data.pivot(index='Cohort Month', columns = ['CohortIndex'], values = 'CustomerID')
cohort_table
```

```python
#Change index
cohort_table.index = cohort_table.index.strftime('%B %Y')
#Turning the cohort table above into a heatmap
plt.figure(figsize=(21,10))
sns.heatmap(cohort_table,annot=True)
```

```python
#Fixing the colour of the cohort table
plt.figure(figsize=(21,10))
sns.heatmap(cohort_table,annot=True,cmap='Blues')
```

```python
#creating a cohort table in percentages
percent_cohort_table = cohort_table.divide(cohort_table.iloc[:,0],axis=0)
percent_cohort_table
```

```python
#Creating a visual of the cohort table in percentages

plt.figure(figsize=(21,10))
sns.heatmap(percent_cohort_table,annot=True,fmt='.0%')
```
